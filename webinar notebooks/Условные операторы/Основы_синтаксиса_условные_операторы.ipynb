Операторы — это некий функционал, производящий какие-либо действия, который может быть представлен в виде символов, например «+», или специальных зарезервированных слов.

Те переменные или выражения, которые сравниваются, как раз и называются операндами.

Операторы могут производить некоторые действия над данными, и эти данные называются операндами.

В общем смысле все операторы делятся на несколько типов. Среди них можно выделить:

* унарные;

    пример: not
    
* бинарные;

    пример: +, >, >=
    
* тернарные.

    пример: a = "Да" if a >= b else "Нет"

## <center> Условный оператор

- это способ управлять выполнением программы;
- это способ запрограммировать принятие решений;
- это логическое выражение после которого пишутся команды, которые выполняются, если условие истинно.

Условная конструкция может включать три части:
- блок if (обязательная часть);
- блоки elif (необязательные, их может быть любое количество);
- else (необязательный блок, может быть один, завершает условную конструкцию).

Итак, условная инструкция в Питоне имеет следующий синтаксис:

```
if Условие 1:
    Блок инструкций 1
elif Условие 2:
    Блок инструкций 2
...
elif Условие n:
    Блок инструкций n
else:
    Блок инструкций n+1
```


Напишем код, который будет определять, является ли число четным

````python
# обращайте внимание на отступы!
x = 82192
if x % 2 == 0:
    print(x, '- четное число')
print('After if')
````

````python
x = 149
if x % 2 == 0:
    print(x, '- четное число')
else:
    print(x, '- нечетное число')
````

````python
# условия выполняются по порядку, как только одно выполнено - проверки останавливаются
parameter_a = 199
if parameter_a < 170:
    print('Class A')
elif parameter_a < 200:
    print('Class C')
else:
    print('Class B')
````

````python
# блоков elif может быть хоть сколько
parameter_a = 175
if parameter_a < 170:
    print('Class A')
elif parameter_a < 180:
    print('Class C')
elif parameter_a < 200:
    print('Class D')
else:
    print('Class B')
````

````python
parameter_a = 30
parameter_b = 180
if parameter_a < 170 and 18 <= parameter_b <= 27:
    print('Class A')
elif parameter_a < 185 and 18 <= parameter_b <= 27:
    print('Class B')
elif parameter_a < 200 and 18 <= parameter_b <= 27:
    print('Class C')
elif parameter_b > 27:
    print('Class D')
else:
    print('Not class')
````

````python
# условия можно вкладывать друг в друга
parameter_a = 170
parameter_b = 18
if 18 <= parameter_b <= 27:
    if parameter_a < 170:
        print('Class A')
    elif parameter_a < 185:
        print('Class B')
    elif parameter_a < 200:
        print('Class C')
    else:
        print('Not class')
else:
    print('Class D')
````

### Задача **"Ход королевы"**:

Шахматный ферзь ходит по диагонали, горизонтали или вертикали. Даны две различные клетки шахматной доски, определите, может ли ферзь попасть с первой клетки на вторую одним ходом. В случае, если хотя бы одно из введенных чисел не лежит в диапазоне от 1 до 8, выведите строку "Ошибка!".

image.png

````python
x_current = 8
y_current = 3
x_new = int(input('Enter new x '))
y_new = int(input('Enter new y '))

if (x_current == x_new) and (y_current == y_new):
    print('Хода нет')
elif abs(x_current - x_new) == abs(y_current - y_new): #ход по диагонали
    print('Успешный ход')
elif x_current == x_new: #ход по вертикали
    print('Успешный ход')
elif y_current == y_new: #ход по горизонтали
    print('Успешный ход') 
else: 
    print('Неуспешный ход')
print('Выполнена попытка переместить ферзя из клетки:', (x_current, y_current), 'в клетку:', (x_new, y_new))
````

## <center> Исключения

**Исключения** — это ошибки, которые выбрасываются при неправильной работе программы и останавливают её выполнение, если они не обработаны.

Для обработки исключений используется конструкция try-except, которая имеет вид:

```
try:
    *код, который может вызвать ту или иную ошибку*
except *ошибка*:
    *код, который выполнится в случае возникновения ошибки*
else:
    *код, который выполнится только в случае, если в try ничего не сломалось*
finally:
    *код, который выполнится в любом случае*
```

Подробнее: https://pythonworld.ru/tipy-dannyx-v-python/isklyucheniya-v-python-konstrukciya-try-except-dlya-obrabotki-isklyuchenij.html

````python
print("Before exception") 
a = 5 #int(input("a: "))
b = 0 #int(input("b: "))
c = a / b
print(c)
````

````python
try:
    print("Before exception") 
    a = 5 #int(input("a: "))
    b = 0 #int(input("b: "))
    c = a / b
    print(c)
except ZeroDivisionError as e:
    print("Zero error", e) 
except IndexError as e:
    print("Index error", e) 
except KeyError as e:
    print("Key error", e)
except Exception as e:
    print("EROR", e) 
else: 
    print("Everything's fine!") 
finally:
    print("Finally finished!") 
````

### Задача "Доступ к конфигурации сервера"

Пользователь пытается обратиться к параметрам конфигурации некоторого сервера. Для этого он указывает последовательность ключей словаря config. Необходимо отловить ошибочный ввод пользователя

````python
config = {
    "server": {
        "host": "127.0.0.1",
        "port": "22"
    },
    "configuration": {
        "ssh": {
            "access": True,
            "login": "some",
            "password": "some"
        }
    }
}

try: 
    keys = input('Введите имена ключей доступа через пробел: ')
    keys_lst = keys.split(' ')
    if len(keys_lst) == 0:
        raise ValueError
    #цикл: перебор последовательности ключей (будем изучать позже)
    value = config
    for key in keys_lst:
        value = value[key]
    print(value)
except ValueError:
    print('Введена пустая строка')
except KeyError:
    print('Неверный формат ключей')
````
