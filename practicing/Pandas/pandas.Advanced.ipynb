```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head(5)
```

```python
display(melb_df.info())
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)

melb_df['Quarter'] = melb_df['Date'].dt.quarter()
```

```python
melb_df['Quarter'].value_counts()
```

```python
melb_df.columns
```

```python
def DT_category(DF, Except):
    
    for col in DF.columns:
        if col not in Except and DF[col].nunique() < 150:
            DF[col] = DF[col].astype('category')
    return DF

Except_list = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

melb_df = DT_category(DF=melb_df, Except=Except_list)
melb_df = melb_df.drop('Quarter', axis=1)
            
```

```python
melb_df.info()
```

Метод sort_values() - Для сортировки значений в DataFrame по значениям одного или нескольких столбцов

        by — имя или список имён столбцов, по значениям которых производится сортировка.
        axis — ось, по которой производится сортировка (0 — строки, 1 — столбцы). По умолчанию сортировка производится по строкам.
        ascending — сортировка по возрастанию (от меньшего к большему). По умолчанию параметр выставлен на True, для сортировки по убыванию (от большего к меньшему) необходимо выставить его на False.
        ignore_index — создаются ли новые индексы в таблице. По умолчанию выставлен на False и сохраняет индексы изначальной таблицы.
        inplace — производится ли замена исходной таблицы на отсортированную. По умолчанию параметр выставлен на False, то есть замены не производится. Чтобы переопределить исходную таблицу на отсортированную, необходимо выставить этот параметр на True.



```python
#отсортируем таблицу сначала по убыванию расстояния от центра города (Distance), а затем — по убыванию цены объекта (Price). 
#Для того чтобы вывод был более наглядным, выделим каждую десятую строку из столбцов Distance и Price результирующей таблицы:
melb_df.sort_values(by=['Distance', 'Price'], ascending=False).loc[::10,['Distance', 'Price']]
```

```python
Метод groupby()

        by — имя или список имён столбцов, по которым производится группировка.
        axis — ось, по которой производится группировка (0 — строки, 1 — столбцы). По умолчанию группировка производится по строкам.
        as_index — добавляется ли дополнительный индекс к таблице. По умолчанию установлен на True.

к этому объекту можно применять уже знакомые нам агрегирующие методы (mean, median, sum и т. д.), чтобы рассчитывать показатели внутри каждой группы.
```

```python
melb_df.groupby(by='Type').mean().loc[:,['Rooms','Price']]
```

```python
melb_df.groupby(by='Type')[['Rooms', 'Price']].mean()
```

```python
melb_df.groupby(by='Regionname')['Distance'].min().sort_values(ascending=False)
```

Группировка данных по одному критерию с несколькими агрегациями
можно воспользоваться методом agg(), который принимает список строк с названиями агрегаций

```python
melb_df.groupby(by='MonthSale')['Price'].agg(['count', 'mean', 'max']).sort_values(by='count', ascending=False)
```

```python
melb_df.groupby('MonthSale')['Price'].agg('describe')
```

```python
melb_df[melb_df['MonthSale'] == 2]['Price'].sort_values(ignore_index=True)[249]
```

```python
melb_df.groupby('Regionname')['SellerG'].agg(
    		['nunique', set]
).sort_values(by='nunique', ascending=False)
```

```python
#Сгруппируйте данные по признаку количества комнат и найдите среднюю цену объектов недвижимости в каждой группе. 
#В качестве ответа запишите количество комнат, для которых средняя цена наибольшая.

melb_df.groupby(by='Rooms', as_index=False)['Price'].mean().sort_values(by='Price', ascending=False, ignore_index=True).loc[0, 'Rooms']
```

```python
melb_df.groupby(by='Regionname', as_index=False)['Lattitude'].std().sort_values(by='Lattitude', ignore_index=True).loc[0, 'Regionname']
```
